```python
X=data.loc[:,data.columns!='class']

y=data['class']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=20,test_size=0.2,stratify=y)
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
# from sklearn.tree import export_graphviz # display the tree within a Jupyter notebook
from IPython.display import SVG
# from graphviz import Source
from IPython.display import display
from ipywidgets import interactive, IntSlider, FloatSlider, interact
import ipywidgets
from IPython.display import Image
from subprocess import call
import matplotlib.image as mpimg
from sklearn.tree import export_graphviz
from graphviz import Source
# conda install python-graphviz
```

```python
X_test
```

```python
#plot tree of decision tree
@interact
def plot_tree(crit=['gini','entropy'],
                 split=['best','random'],
                 depth=IntSlider(min=1,max=30,value=2, continuous_update=False),#value is default
                 min_split=IntSlider(min=2,max=5,value=2, continuous_update=False),#min values required for split
                 min_leaf=IntSlider(min=1,max=5,value=1, continuous_update=False)):#min values to be present in leaf
    estimator=DecisionTreeClassifier(random_state=42,criterion=crit,splitter=split,max_depth=depth,
                                     min_samples_split=min_split,
                                     min_samples_leaf=min_leaf)
    estimator.fit(X_train,y_train)
    print("decision tree accuracy is {:.3f}".format(accuracy_score(y_train,estimator.predict(X_train))))
    print("decision tree accuracy is {:.3f}".format(accuracy_score(y_test,estimator.predict(X_test))))
    graph=Source(tree.export_graphviz(estimator, out_file=None,
                                     feature_names=X_train.columns,
                                     class_names=['Disliked','Liked'],
                                     filled=True))
    display(Image(data=graph.pipe(format='png')))
    return estimator
    
```

```python
#plot tree of random forest
@interact
def plot_tree_rf(crit=['gini','entropy'],
                 bootstrap=['best','random'],
                 depth=IntSlider(min=1,max=30,value=3, continuous_update=False),#value is default
                 forests=IntSlider(min=1,max=200,value=100,continuous_update=False),
                 min_split=IntSlider(min=2,max=5,value=2, continuous_update=False),#min values required for split
                 min_leaf=IntSlider(min=1,max=5,value=1, continuous_update=False)):#min values to be present in leaf
    estimator=RandomForestClassifier(random_state=42,criterion=crit,bootstrap=bootstrap,
                                     n_estimators=forests,
                                     max_depth=depth,
                                     min_samples_split=min_split,
                                     min_samples_leaf=min_leaf,
                                    n_jobs=-1,
                                    verbose=False)
    estimator.fit(X_train,y_train)
    print("Random Forest accuracy is {:.3f}".format(accuracy_score(y_train,estimator.predict(X_train))))
    print("Random Forest accuracy is {:.3f}".format(accuracy_score(y_test,estimator.predict(X_test))))
    
    #printing tree 1
    num_tree=estimator.estimators_[0]
    graph=Source(tree.export_graphviz(num_tree,out_file=None,
                                     feature_names=X_train.columns,
                                     class_names=['Disliked','Liked'],
                                     filled=True))
    display(Image(data=graph.pipe(format='png')))
    return estimator
    
```
